# Plot workspace

### Imports

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

import torch
import torch.nn as nn

# set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
```

### Load files

```python
episode_rewards = torch.load("episode_rewards.kek")
```

### Ploting

```python
rewards_t = torch.tensor(episode_rewards, dtype=torch.float)
plt.title('Reward over Episodes')
plt.xlabel('Episode')
plt.ylabel('Reward')
#plt.plot(episode_durations, label="durations")
plt.scatter(range(rewards_t.numpy().shape[0]), rewards_t.numpy(), label="reward")
    
# Take 100 episode averages and plot them too
if len(rewards_t) >= 100:
    means = rewards_t.unfold(0, 100, 1).mean(1).view(-1)
    means = torch.cat((torch.zeros(99), means))
    plt.plot(means.numpy(), label="reward mean", color="orange")
plt.legend()
plt.show()
```
